```python
import pandas as pd
import numpy as np
import pickle
import tensorflow as tf
from tensorflow.keras.models import load_model
```

```python
# Load the model

model=load_model('model.h5')

 # Load the pickle files now

with open('scaler.pkl','rb') as file:
    scaler=pickle.load(file)

with open('onehot_encoder_geo.pkl','rb') as file:
    onehot_encoder_geo=pickle.load(file)

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender=pickle.load(file)


 
 
```

```python
# Example input data

input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}

input=pd.DataFrame([input_data])
input
```

```python
input['Gender']=label_encoder_gender.transform(input['Gender'])
encoded_geo=onehot_encoder_geo.transform(input[['Geography']]).toarray()
col=pd.DataFrame(encoded_geo,columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
input=pd.concat([input.drop(columns=['Geography']),col],axis=1)
input_scaled=scaler.transform(input)
```

```python
input_scaled
```

```python
prediction=model.predict(input_scaled)
```

```python
prediction_prob=prediction[0][0]
```

```python
if prediction_prob>0.5:
    print(f"The person will churn/exit")
else:
    print("The person will now churn/exit")
```
